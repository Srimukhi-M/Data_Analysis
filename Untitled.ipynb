```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv(r"C:\Users\Mekarthi Srimukhi\Desktop\amazon.csv")
```

```python
category_ratings = df.groupby('category')['rating'].mean().reset_index()
plt.bar(category_ratings['category'], category_ratings['rating'])
plt.xlabel('Category')
plt.ylabel('Avg Rating')
plt.show()
```

```python
top_products = df.sort_values(by='rating', ascending=False).head(5)
bottom_products = df.sort_values(by='rating').head(5)
print(top_products[['product_id', 'rating']])
print(bottom_products[['product_id', 'rating']])
```

```python

sns.histplot(df['actual_price'], bins=20)
plt.title('Frequency Distribution  of Actual Prices')
plt.xlabel('Actual Price')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
sns.histplot(df['discounted_price'], bins=20)
plt.title('Frequency Distribution  of discounted Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
